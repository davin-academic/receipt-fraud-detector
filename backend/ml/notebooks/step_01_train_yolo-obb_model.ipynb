# Step 01 of 04 - Yolo26-OBB model training.
### Train object detection model to extract receipt features.

```python
from ultralytics import YOLO
import torch
from pathlib import Path

print(f"CUDA available: {torch.cuda.is_available()}")
print(f"Device: {torch.cuda.get_device_name(0) if torch.cuda.is_available() else 'CPU'}")
```

**Run notebook kernel from backend/ml/notebooks or update ML_DIR path.**

```python
ML_DIR = Path.cwd().parent
yaml_path = ML_DIR / "stage_01.yaml"

model = YOLO('yolo26l-obb')
```

**As our dataset was very small we augmented the dataset during training to improve model performance.**

```python
results = model.train(
    data=yaml_path,
    epochs=200,
    imgsz=1024,
    batch=16,
    device=0,
    patience=30,
    workers=8,
    
    # Data augmentation
    mosaic=1.0,
    close_mosaic=15,
    mixup=0.15,
    hsv_h=0.015,
    hsv_s=0.5,
    hsv_v=0.3,
    degrees=15,           
    translate=0.1,
    scale=0.4,
    fliplr=0.0,       
    flipud=0.0,
    
    optimizer='auto',
    lr0=0.01,
    
    project='runs/stage1',
    name='date_detector_obb_v1',
    save=True,
    plots=True,
)

print(f"\nBest weights: {results.save_dir}/weights/best.pt")
```
